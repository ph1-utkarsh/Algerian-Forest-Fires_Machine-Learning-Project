```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

DATA CLEANING

```python
##msiing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index
1. "Bejai Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df = dataset
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns=df.columns.str.strip()
```

```python
df.info()
```

Change the Datatype of Columns 

```python
# day , month , year , temp. ,RH ,Ws into integers
df[["day" , "month" , "year" , "Temperature" ,"RH" ,"Ws"]] = df[["day" , "month" , "year" , "Temperature" ,"RH" ,"Ws"]].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
##changing other columns into float data type
df[["Rain" ,"FFMC","DMC","DC","ISI","BUI","FWI"]]=df[["Rain" ,"FFMC","DMC","DC","ISI","BUI","FWI"]].astype(float)
```

```python
df.info()
```

```python
df.head()
```

Save the Cleaned Dataset

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

Exploratory Data Analysis

```python
df_copy = df.drop(['day','year','month'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy["Classes"].value_counts()
```

```python
##pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
class_labels = ["Fire" ,"Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage, labels = class_labels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```

```python
##Correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),'not fire','fire')
```

```python
# Monthly Fire Analysis for Sidi-bel Regions and Bajaia Regions
fig, axes = plt.subplots(2, 1, figsize=(13, 12))  # Create a 2x1 grid for subplots

# Sidi-bel Regions Analysis
dftemp_sidi_bel = df.loc[df['Region'] == 1]
sns.set_style("whitegrid")
sns.countplot(ax=axes[0], x='month', hue='Classes', data=dftemp_sidi_bel)
axes[0].set_ylabel('Number of Fires', weight='bold')
axes[0].set_xlabel('Months', weight='bold')
axes[0].set_title("Fire Analysis of Sidi-bel Regions", weight='bold')

# Bajaia Regions Analysis
dftemp_bajaia = df.loc[df['Region'] == 0]
sns.set_style("whitegrid")
sns.countplot(ax=axes[1], x='month', hue='Classes', data=dftemp_bajaia)
axes[1].set_ylabel('Number of Fires', weight='bold')
axes[1].set_xlabel('Months', weight='bold')
axes[1].set_title("Fire Analysis of Bajaia Regions", weight='bold')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
